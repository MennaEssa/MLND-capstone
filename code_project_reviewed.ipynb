Adding preprocessing function for easier use , it selects the features selected fields , scales with with standard scaler and finally return the new features set and labels.

```python
from sklearn.feature_selection import SelectPercentile , f_regression

def preprocess(filename):
    data=pd.read_csv(filename).drop('Name' , axis=1)
    labels=data['Malware']
    features = data[['e_maxalloc', 'e_ip' ,'e_oemid' ,'e_oeminfo', 'e_lfanew', 'Machine',
                    'NumberOfSections', 'TimeDateStamp' ,'NumberOfSymbols',
                     'SizeOfOptionalHeader', 'Characteristics', 'Magic', 'MajorLinkerVersion',
                     'MinorLinkerVersion', 'AddressOfEntryPoint' ,'ImageBase' ,'SectionAlignment',
                    'FileAlignment', 'MajorSubsystemVersion', 'CheckSum' ,'Subsystem',
                    'DllCharacteristics', 'SizeOfStackReserve', 'SizeOfHeapReserve',
                     'SuspiciousImportFunctions' ,'SuspiciousNameSection', 'SectionsLength',
                     'SectionMinEntropy' ,'SectionMinRawsize' ,'SectionMinVirtualsize',
                     'SectionMaxPointerData' ,'SectionMaxChar' ,'DirectoryEntryImport',
                    'DirectoryEntryExport']]

    scaler = StandardScaler()
    scale_np = scaler.fit_transform(features)
    features_scaled = pd.DataFrame(scale_np, columns=features.columns)
    display(features_scaled.head(n=1))
    print("post features selection data" , features_scaled.shape)
    return features_scaled , labels

    
```

```python
features_processed  , labels= preprocess('dataset_malwares.csv')
X_train, X_test, y_train, y_test = train_test_split(features_processed , labels , test_size=0.25, random_state=0 , shuffle=True )
```

<h3> Models training </h3>

```python
def pretty_print(results):
    '''
    print the training results dictionary 
    input : results dict
    '''
    
    for classifier in results:
        print(classifier)
        for item in results[classifier]:
            print(item , ":" , results[classifier][item])
```

```python
from sklearn.metrics import fbeta_score,roc_auc_score
from time import time
from sklearn.metrics import confusion_matrix
import print_cm

def train_predict(learner, X_train, y_train, X_test, y_test): 
    '''
    inputs:
       - learner: the learning algorithm to be trained and predicted on
       - X_train: features training set
       - y_train: label training set
       - X_test: features testing set
       - y_test: label testing set
    '''
    results = {}
    learner.fit(X_train,y_train)
    predictions_test =learner.predict(X_test)
    predictions_train = learner.predict(X_train[:300])    
    results['auc_train'] = roc_auc_score(y_train[:300], predictions_train)
    results['auc_test'] = roc_auc_score(y_test, predictions_test)
    results['f_train'] = fbeta_score(y_train[:300] , predictions_train , beta=0.5)
    results['f_test'] = fbeta_score(y_test , predictions_test , beta=0.5)       
    print("{} trained on samples.".format(learner.__class__.__name__))
    cm = confusion_matrix(y_test, predictions_test)
    df_cm = pd.DataFrame(cm)
    print_cm.pretty_plot_confusion_matrix(df_cm, cmap='PuRd')
    return results
```

```python
from sklearn.naive_bayes import GaussianNB
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import RandomizedSearchCV
from sklearn.model_selection import ShuffleSplit
from sklearn.metrics import make_scorer
from xgboost import XGBClassifier
import visuals as vs

#Grid search for Random forest
regressor = RandomForestClassifier()
cv_sets = ShuffleSplit(n_splits = 10, test_size = 0.25, random_state = 0)
params = {'n_estimators':[100,150,200] , 
          'max_depth': range(10,20) ,
          'max_features' : ['sqrt' , 'auto', 'log2']}
scoring_fnc =  make_scorer(fbeta_score , beta=0.5)
grid = GridSearchCV(regressor,params,scoring_fnc,cv=cv_sets , n_jobs=-1 , verbose=10)
grid = grid.fit(X_train, y_train)

print(grid.best_params_)
```
